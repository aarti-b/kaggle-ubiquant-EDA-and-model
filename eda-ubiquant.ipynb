```python
X_features = [f'f_{i}' for i in range(300)]
Y = 'target'
```

```python
from sklearn.preprocessing import StandardScaler

x = data[:10000].loc[:, X_features].values
x = StandardScaler().fit_transform(x)
pca = PCA(n_components=300)
principalComponents = pca.fit_transform(x)

```

```python
principal_comp_ubiquent = pd.DataFrame(data = principalComponents)
```

```python
principal_comp_ubiquent.head()
```

```python
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
plt.figure()
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Ubiquent",fontsize=20)
sb.scatterplot(data=principal_comp_ubiquent, x=principal_comp_ubiquent.loc[0], y=principal_comp_ubiquent.loc[1], color='r')
```

```python
'''
Non stationarity check 
'''
'''
p-value <0.05 null hypothesis is rejected ie the series is stationary ...
'''


from statsmodels.tsa.stattools import adfuller

X = data['target'][:80000].values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
'''
ADF in series from investment ID = 1
'''
Y = filtered_id_1['target'].values
result = adfuller(Y)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

```
